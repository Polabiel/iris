```python

# Importação de bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Importação do dataset
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# importação do KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
# Carregamento do dataset
dataset = load_iris()
```

```python
# Atributos dos dataset principais
# x_1 = Comprimento da perola, x_2 = largura da perola, x_3 comprimeto sepula, x_4 largura da sepula
x = pd.DataFrame(dataset['data'], columns = dataset['feature_names'])

# y = target = 0, 1, 2 (setosa, versicolor, virginica)
y = pd.Series(dataset['target'], name = 'target')
```

```python
# criação da figura
plt.figure(4, figsize= (10,8))

# Definindo constantes para os literais duplicados
SEPAL_LENGTH = 'sepal length (cm)'
SEPAL_WIDTH = 'sepal width (cm)'

# Gráfico de dispersão de cada espécie
plt.scatter(x[SEPAL_LENGTH][:50], x[SEPAL_WIDTH][:50], label='Iris Setosa')
plt.scatter(x[SEPAL_LENGTH][50:100], x[SEPAL_WIDTH][50:100], label='Iris Versicolor')
plt.scatter(x[SEPAL_LENGTH][100:], x[SEPAL_WIDTH][100:], label='Iris Virginica')
plt.xlabel('Comprimento da sépala (cm)', fontsize=20)
plt.ylabel('Comprimento vs. Largura da sépala')
plt.legend(prop = {'size': 18})
plt.show()
```

```python
# Divisão do dataset em treino e teste
x_train, x_test, y_train, y_test = train_test_split(x ,y ,test_size=0.50, train_size=0.50)
```

```python
# Criação do modelo
model = Pipeline(steps=[
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier(n_neighbors=5))
])
```

```python
# Treinamento do modelo
model.fit(x_train, y_train)
```

```python
# Coletar a Acuácia do modelo
accuracy = model.score(x_test, y_test)

# Exibir a acurácia
print(f'Acurácia: {accuracy:.2f}')
```

```python
train_score = model.score(x_train, y_train)
print( f'Train score: {train_score:.2f}')
test_score = model.score(x_test, y_test)
print(f'Test score: {test_score:.2f}')
```
